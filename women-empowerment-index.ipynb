```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/women-empowerment-index/women_empowerment_index.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
for col in data.select_dtypes(include='object').columns:
    print(col, data[col].unique())
```

```python
# Visualization 1: Distribution of Women's Empowerment Index (WEI) - 2022
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Women\'s Empowerment Index (WEI) - 2022', bins=20, kde=True)
plt.title('Distribution of Women\'s Empowerment Index (WEI) - 2022')
plt.xlabel('WEI - 2022')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Boxplot of Women's Empowerment Index (WEI) - 2022 by Women's Empowerment Group
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Women\'s Empowerment Group - 2022', y='Women\'s Empowerment Index (WEI) - 2022')
plt.title('Boxplot of Women\'s Empowerment Index (WEI) - 2022 by Women\'s Empowerment Group')
plt.xlabel('Women\'s Empowerment Group - 2022')
plt.ylabel('WEI - 2022')
plt.show()
```

```python
# Visualization 3: Scatter plot of Women's Empowerment Index (WEI) - 2022 vs. Global Gender Parity Index (GGPI) - 2022
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Women\'s Empowerment Index (WEI) - 2022', y='Global Gender Parity Index (GGPI) - 2022')
plt.title('Scatter plot of WEI - 2022 vs. GGPI - 2022')
plt.xlabel('WEI - 2022')
plt.ylabel('GGPI - 2022')
plt.show()
```

```python
# Visualization 4: Count plot of Gender Parity Group - 2022
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
ax = sns.countplot(data=data, x='Gender Parity Group - 2022')

plt.title('Count plot of Gender Parity Group - 2022')
plt.xlabel('Gender Parity Group - 2022')
plt.ylabel('Count')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

```python
# Visualization 5: Boxplot of Women's Empowerment Index (WEI) - 2022 by Human Development Group
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Human Development Group - 2021', y='Women\'s Empowerment Index (WEI) - 2022')
plt.title('Boxplot of WEI - 2022 by Human Development Group')
plt.xlabel('Human Development Group - 2021')
plt.ylabel('WEI - 2022')
plt.show()
```

```python
# Visualization 6: Violin plot of Women's Empowerment Index (WEI) - 2022 by Sustainable Development Goal regions
plt.figure(figsize=(12, 8))
sns.violinplot(data=data, x='Sustainable Development Goal regions', y='Women\'s Empowerment Index (WEI) - 2022')
plt.title('Violin plot of WEI - 2022 by Sustainable Development Goal regions')
plt.xlabel('Sustainable Development Goal regions')
plt.ylabel('WEI - 2022')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 7: Pairplot of Women's Empowerment Index (WEI) - 2022, Global Gender Parity Index (GGPI) - 2022,
# and Human Development Group - 2021
sns.pairplot(data=data, vars=['Women\'s Empowerment Index (WEI) - 2022', 'Global Gender Parity Index (GGPI) - 2022'],
             hue='Human Development Group - 2021', height=4)
plt.suptitle('Pairplot of WEI - 2022, GGPI - 2022, and Human Development Group - 2021', y=1.02)
plt.show()
```

```python
# Visualization 8: Barplot of Mean Women's Empowerment Index (WEI) - 2022 by Gender Parity Group - 2022
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Gender Parity Group - 2022', y='Women\'s Empowerment Index (WEI) - 2022', estimator=np.mean)
plt.title('Mean WEI - 2022 by Gender Parity Group - 2022')
plt.xlabel('Gender Parity Group - 2022')
plt.ylabel('Mean WEI - 2022')
plt.show()
```

```python
# Visualization 9: Stacked barplot of Women's Empowerment Group - 2022 by Sustainable Development Goal regions
plt.figure(figsize=(12, 8))
sns.countplot(data=data, x='Sustainable Development Goal regions', hue='Women\'s Empowerment Group - 2022')
plt.title('Women\'s Empowerment Group - 2022 by Sustainable Development Goal regions')
plt.xlabel('Sustainable Development Goal regions')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Women\'s Empowerment Group - 2022')
plt.show()
```

```python
# Visualization 10: Pairplot of Women's Empowerment Index (WEI) - 2022 with hue as Gender Parity Group - 2022
sns.pairplot(data=data, vars=['Women\'s Empowerment Index (WEI) - 2022', 'Global Gender Parity Index (GGPI) - 2022'],
             hue='Gender Parity Group - 2022', height=4)
plt.suptitle('Pairplot of WEI - 2022 and GGPI - 2022 with hue as Gender Parity Group - 2022', y=1.02)
plt.show()
```

```python
# Visualization 11: Heatmap of correlation matrix
# Filter out non-numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr_matrix = numeric_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap of correlation matrix')
plt.show()
```

```python
# Visualization 12: Boxplot of Women's Empowerment Index (WEI) - 2022 by Gender Parity Group - 2022
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Gender Parity Group - 2022', y='Women\'s Empowerment Index (WEI) - 2022')
plt.title('Boxplot of WEI - 2022 by Gender Parity Group - 2022')
plt.xlabel('Gender Parity Group - 2022')
plt.ylabel('WEI - 2022')
plt.show()
```
